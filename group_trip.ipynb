```python
def getSInNum(x):
    if(x=='Winter'):
        return 1
    elif(x=='Spring'):
        return 2
    elif(x=='Summer'):
        return 3
    elif(x=='Fall'):
        return 4
df["season_num"] = df["season"].apply(getSInNum)
```

```python
start = df.groupby(['k_date', 'start_station_name']).agg(count=('start_station_name', 'size'), 
                                                         uvIndex=('uvIndex', 'mean'),
                                                         tempC=('h-tempC', 'mean'),
                                                         totalSnow_cm=('totalSnow_cm', 'mean'),
                                                         mintempC=('mintempC', 'mean'),
                                                         maxtempC=('maxtempC', 'mean'),
                                                         windSpeedKmph=('h-windSpeedKmph', 'mean'),
                                                         precipMM=('h-precipMM', 'mean'),
                                                         humidity=('h-humidity', 'mean'),
                                                         visibilityKM=('h-visibilityKM', 'mean'),
                                                         cloudcover=('h-cloudcover', 'mean'),
                                                         FeelsLikeC=('h-FeelsLikeC', 'mean'),
                                                         is_holiday=('is_holiday', 'mean'),
                                                         day_of_week=('day_of_week', 'mean'),
                                                         season_num=('season_num', 'mean')
                                                        ).reset_index()
```

```python
minim = df.groupby(['k_date']).agg(count=('k_date', 'size'), 
                                                         uvIndex=('uvIndex', 'mean'),
                                                         tempC=('h-tempC', 'mean'),
                                                         totalSnow_cm=('totalSnow_cm', 'mean'),
                                                         mintempC=('mintempC', 'mean'),
                                                         maxtempC=('maxtempC', 'mean'),
                                                         windSpeedKmph=('h-windSpeedKmph', 'mean'),
                                                         precipMM=('h-precipMM', 'mean'),
                                                         humidity=('h-humidity', 'mean'),
                                                         visibilityKM=('h-visibilityKM', 'mean'),
                                                         cloudcover=('h-cloudcover', 'mean'),
                                                         FeelsLikeC=('h-FeelsLikeC', 'mean'),
                                                         is_holiday=('is_holiday', 'mean'),
                                                         day_of_week=('day_of_week', 'mean'),
                                                         season_num=('season_num', 'mean')
                                                        ).reset_index()
```

```python
end = df.groupby(['k_date', 'end_station_name']).agg(count=('end_station_name', 'size'), 
                                                         uvIndex=('uvIndex', 'mean'),
                                                         tempC=('h-tempC', 'mean'),
                                                         totalSnow_cm=('totalSnow_cm', 'mean'),
                                                         mintempC=('mintempC', 'mean'),
                                                         maxtempC=('maxtempC', 'mean'),
                                                         windSpeedKmph=('h-windSpeedKmph', 'mean'),
                                                         precipMM=('h-precipMM', 'mean'),
                                                         humidity=('h-humidity', 'mean'),
                                                         visibilityKM=('h-visibilityKM', 'mean'),
                                                         cloudcover=('h-cloudcover', 'mean'),
                                                         FeelsLikeC=('h-FeelsLikeC', 'mean'),
                                                         is_holiday=('is_holiday', 'mean'),
                                                         day_of_week=('day_of_week', 'mean'),
                                                         season_num=('season_num', 'mean')
                                                        ).reset_index()
```

```python
def getYear(x):
    return int(x.split("-")[2])
#start["year"] = start.k_date.apply(getYear)
#end["year"] = end.k_date.apply(getYear)
minim["year"] = minim.k_date.apply(getYear)
```

```python
def getDay(x):
    spl = x.split("-")
    return spl[0]+"-"+spl[1]
#start["day"] = start.k_date.apply(getDay)
#end["day"] = end.k_date.apply(getDay)
minim["day"] = minim.k_date.apply(getDay)
```

```python
def getMonth(x):
    return int(x.split("-")[0])
#start["month"] = start.k_date.apply(getMonth)
#end["month"] = end.k_date.apply(getMonth)
minim["month"] = minim.k_date.apply(getMonth)
```
